````
# import plotly.plotly is depriciated 
import chart_studio.plotly as py
from plotly.offline import iplot
import plotly.graph_objects as go
import numpy as np
````

````
# Setup x and y randome numbers
N = 300
example_x = np.random.rand(N)
example_y = np.random.rand(N)
````

There are many diffrent way of doing plotly and this is one of them using iplot

```
# graph object is a key value paiir dictionary
trace1 = go.Scatter(
    x = example_x,
    y = example_y,
    mode = 'markers'
)
# Now using trace plot the graph
fig1 = [trace1]
iplot(fig1)
```
However this method is hard to update the plot later time point

````
# Use Figure() function then add trace later
fig01 = go.Figure()
fig01.add_trace(go.Scatter(
    x = example_x, 
    y = example_y, 
    mode = 'markers'));
````

````
# So that fig instance can be updated for layouts
fig01.update_layout(
    autosize=False,
    width=800,
    height=500,
    margin=dict(t=30, b=30, l=50, r=50, pad=4),
    xaxis= go.layout.XAxis(linecolor = 'black',
                          linewidth = 2,
                          mirror = False),

    yaxis= go.layout.YAxis(linecolor = 'black',
                          linewidth = 2,
                          mirror = False),
    paper_bgcolor="LightSteelBlue",
)

fig01.show()
````

# 2. Pie Chart
A) Reference
Plotly documentation and user API is wonderful. There are so many code examples. <br>
https://plotly.com/python/pie-charts/<br>



````
# Break down my expense per month
groups = ['Rent', 'Food', 'Internet', 'Mobile', 'Car_Insurance', 'Gas', 'Interest', 'Misc' ]
# Creat amount of each categories
amounts = [900, 400, 70, 110, 300, 80, 250, 100]
# Style
colors = ['#32a844', '#90a832', '#15b7d4', '#1578d4', '#d48b15', '#d415c7', '#d41532', '#d41532']
````

````
# Define trace here
trace02 = go.Pie(labels=groups, 
                 values=amounts, 
                 hoverinfo='label+percent', 
                 textinfo='value', 
                 textfont=dict(size=25), 
                 marker=dict(colors=colors, 
                             line=dict(color='#000000', 
                                       width=1)))

# Define layout" Notice Layout " 
layout02 = go.Layout(
    autosize=False,
    width=500,
    height=500,
    title = 'My basic expenses')

# Using both trace and Layout plot the Pie chart
fig02 = go.Figure(data = trace02, layout = layout02)
iplot(fig02, show_link = False)
````
