````
# Note, github raw adress has to be used
# url01 = 'https://github.com/kingdavid72/plotly_examples_choi/blob/master/01-plotly/gamerdata02.csv'
url= 'https://raw.githubusercontent.com/kingdavid72/plotly_examples_choi/master/01-plotly/gamerdata02.csv'

# Since the data has faulty row got to be skipped
# and also use header=0 not header=None
df_ori = pd.read_csv(url, error_bad_lines=False, header=0)
df_num = df_ori.apply(pd.to_numeric, errors='ignore')
df_num.head()
````

## 2. Test out graphing using scatter

````
print(df_new.columns.tolist())
````

````
# mode='makers' removes line
trace01 = go.Scatter(x=df_new['solo_RoundsPlayed'], 
                     y=df_new['solo_Wins'],
                     name='Rounds Won',
                     mode = 'markers')
# checkout the trace01
# print(trace01)

# it is stupid that Layout doesn't have width and height and showlegend moved to update_layout
layout01 = go.Layout(title='PUBG Wins vs Rounds Played', 
                     plot_bgcolor='rgb(230, 230, 230)')

# Generate fig01 using trace and layout
fig01 = go.Figure(data=trace01, layout=layout01)

# To customizae width and height
fig01.update_layout(
    autosize=False,
    width=500,
    height=300,
    margin=dict(t=50, b=30, l=50, r=50, pad=4),
    xaxis= go.layout.XAxis(linecolor = 'black',
                          linewidth = 2,
                          mirror = False),

    yaxis= go.layout.YAxis(linecolor = 'black',
                          linewidth = 2,
                          mirror = False),
    paper_bgcolor="LightSteelBlue",
    showlegend=True
)

# Then finally generate figure
iplot(fig01, filename='gamewindata.png')
````

## 3. Bar Chart

````
df_bar = df_num.head(30)

trace02 = go.Bar(x=df_bar['player_name'],
                 y=df_bar['solo_RoundsPlayed'],
                 name='Round Played')

trace03 = go.Bar(x=df_bar['player_name'],
                 y=df_bar['solo_Wins'],
                 name='Wins')

data02 = [trace02, trace03]
layout02 = go.Layout(barmode='group')

fig02 = go.Figure(data=data02, layout=layout02)
fig02.update_layout(
    autosize=False,
    width=1000,
    height=500,
    margin=dict(t=50, b=30, l=50, r=50, pad=4),
    xaxis= go.layout.XAxis(linecolor = 'black',
                          linewidth = 2,
                          mirror = False),

    yaxis= go.layout.YAxis(linecolor = 'black',
                          linewidth = 2,
                          mirror = False),
    paper_bgcolor="LightSteelBlue",
    showlegend=True
)
iplot(fig02)
````

# B. Box & Whisker Plot

````
x_data = ['Carmelo Anthony', 'Dwyane Wade',
          'Deron Williams', 'Brook Lopez',
          'Damian Lillard', 'David West',]

y0 = np.random.randn(50)-1
y1 = np.random.randn(50)+1
y2 = np.random.randn(50)
y3 = np.random.randn(50)+2
y4 = np.random.randn(50)-2
y5 = np.random.randn(50)+3

y_data = [y0,y1,y2,y3,y4,y5]

colors = ['rgba(93, 164, 214, 0.5)', 'rgba(255, 144, 14, 0.5)', 'rgba(44, 160, 101, 0.5)', 'rgba(255, 65, 54, 0.5)', 'rgba(207, 114, 255, 0.5)', 'rgba(127, 96, 0, 0.5)']
````

````
trace04 = []

for xd, yd, cls in zip(x_data, y_data, colors):
        trace04.append(go.Box(
            y=yd,
            name=xd,
            boxpoints='all',
            jitter=0.5,
            whiskerwidth=0.2,
            fillcolor=cls,
            marker=dict(
                size=2,
            ),
            line=dict(width=1),
        ))

layout04 = go.Layout(
    autosize=False,
    width=1000,
    height=500,
    title='Points Scored by the Top 9 Scoring NBA Players in 2012',
    yaxis=dict(
        autorange=True,
        showgrid=True,
        zeroline=True,
        dtick=5,
        gridcolor='rgb(255, 255, 255)',
        gridwidth=1,
        zerolinecolor='rgb(255, 255, 255)',
        zerolinewidth=2,
    ),
    margin=dict(
        l=40,
        r=30,
        b=80,
        t=100,
    ),
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
    showlegend=False
)

fig04 = go.Figure(data=trace04, layout=layout04)
iplot(fig04)
````
