```
# Creat data
x = df_new['solo_Wins']
y = df_new['solo_TimeSurvived']
#z = df_new['squad_Heals']

# Set the color scale
colorscale = ['#7A4579', '#D56073', 'rgb(236,158,105)', (1, 1, 0.2), (0.98,0.98,0.98)]

# Generate 2D density plot
fig01 = ff.create_2d_density(x, y, 
                             colorscale=colorscale, 
                             hist_color='rgb(255, 237, 222)', 
                             point_size=4,
                             height=500, width=800)

# Finally draw
iplot(fig01, filename='histogram_subplots')
```

## 2. 3D Plot visulization
1) 3D plot Plotly Documentation <br>
https://plotly.com/python/3d-axes/<br>
what a stupid here only in 3D Plot axis tiltle should be on scene. Do not try to assign directtly like 2D scatter.


```
# Setup the data
x = df_new['solo_Wins']
y = df_new['solo_TimeSurvived']
z = df_new['solo_RoundsPlayed']

# Asssign trace
trace02 = go.Scatter3d(
                        x=x,
                        y=y,
                        z=z,
                        mode='markers',
                        marker=dict(
                        size=12,
                        color=z,
                        colorscale='Viridis',
                        opacity=0.8))
# Creating data array
data02 = [trace02]

# setup layout Notice axis title is on scene)
layout02 = go.Layout(scene = dict(
                    xaxis_title='Num. of Wins',
                    yaxis_title='Time Survived (min)',
                    zaxis_title='Num. of Played'),
                    height=600,
                    width=900,
                    margin=dict(t=0,b=0, l=0, r=0)
                   )

# Generate the figure
fig02 = go.Figure(data=data02, layout=layout02)

# To be able to edit online use py.iplot() instead iplot()
iplot(fig02, filename='3d-pubg-plot')
# py.iplot(fig02, filename='3d-pubg-plot')
```
